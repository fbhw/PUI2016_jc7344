```python
from __future__ import print_function
__author__= 'jc'

import numpy as pn 
import pylab as pl
import matplotlib.pyplot as pt
import plotly.plotly as py
from scipy.optimize import curve_fit, minimize
    
%pylab inline
```

```python
mysize = (1500 /np.array(range(1, 101)))
print (mysize)
```

```python
dist = ['chisq']
```

```python
assigned_mean = 50
```

```python
df = assigned_mean
```

```python
dic = {}
dic['chisq'] = np.random.chisquare(df, size = 100)
pl.hist(dic['chisq'], bins = 30)
pl.ylabel('S')
pl.xlabel('C')
```

```python
print (dic['chisq'].mean())
```

```python
dic['chisq'] = {} 


for n in mysize:
    dic['chisq'][n] = np.random.chisquare(df, size = n)
    dic['chisq']['means'] = {}
newchisq = pl.figure(figsize=(10,6)).add_subplot(111)
for nn in dic['chisq']:
    if not type(nn) == str:
        dic['chisq']['means'][nn] = dic['chisq'][nn].mean()
        
        newchisq.plot(nn, dic['chisq']['means'][nn], 'o')
        newchisq.set_xlabel('sample size', fontsize=18)
        newchisq.set_ylabel('sample mean', fontsize=18)
        newchisq.set_title('Chi squared', fontsize=18)
        newchisq.plot([min(mysize), max(mysize)], [df, df], 'k')
```
